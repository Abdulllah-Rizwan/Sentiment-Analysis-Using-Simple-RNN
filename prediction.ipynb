```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.datasets import imdb
```

```python
word_index = imdb.get_word_index()
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
reverse_word_index
```

```python
model = load_model('model.h5')

model.summary()
```

```python
model.get_weights()
```

```python
def decode_review(encoded_review):
    return ' '.join(reverse_word_index.get(i-3,'?') for i in encoded_review)

def process_text(text):
    words = text.lower().split()
    encoded_review = [ word_index.get(word,2)+3 for word in words ]
    padded_review = sequence.pad_sequences([encoded_review], maxlen=250)
    return padded_review
```

```python
def predict_sentiment(review):
    preprocessed_input = process_text(review)
    
    pred = model.predict(preprocessed_input)
    
    sentiment = 'Positive' if pred[0][0] > 0.5 else 'Negative'
    
    return sentiment,pred[0][0]
```

```python
example_review = "The movie was a letdown. The special effects were subpar, the story was convoluted, and it lacked any real emotional depth. I wouldn't watch it again."


sentiment,probability = predict_sentiment(example_review)

print(f'Sentiment: {sentiment}\nProbability: {probability}')
```
